```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list,"y") # y , yellow anlamına geliyor. g,b,r yazabiliriz
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#works with numpy
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"r")
plt.xlabel(np_age_list)
plt.ylabel(np_weight_list)
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20) 
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 **3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2)
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2,numpy_arr1,"g--")
plt.show()
```

```python
#customize the axes #eksenleri özelleştirmek
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.7,0.7])
figure_axes.plot(numpy_arr1,numpy_arr2,"r")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
```

```python
plt.show()
```

```python
#customize labels
```

```python
new_figure = plt.figure(dpi = 100)

new_axes = new_figure.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1,numpy_arr1 ** 2,label = "numpy array ** 2")
new_axes.plot(numpy_arr1,numpy_arr1 ** 3 , label = "numpy array **3")

new_axes.legend(loc = 6)
```

```python
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig,my_axes = plt.subplots()
```

```python
type(my_fig)
```

```python
type(my_axes)
```

```python
my_fig,my_axes = plt.subplots()

my_axes.plot(data1,data2,color = "#3A95A8",alpha = 0.3)
my_axes.plot(data2,data1,color = "#C96F23")
```

```python
plt.show()
```

```python

(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color="blue", linewidth = 1.0)
new_axes.plot(data1, data1 + 3, color ="yellow", linewidth = 1.0)

new_axes.plot(data1, data1 + 4, color = "#C96F23", linestyle = "-.")
new_axes.plot(data1, data1 + 5, color = "#C96F23", linestyle = ":")
new_axes.plot(data1, data1 + 6, color = "#C96F23", linestyle = "--")

new_axes.plot(data1, data1 + 7, color = "#000000", linestyle = "--", marker = "o",markersize = 8, markerfacecolor="red")
new_axes.plot(data1, data1 + 8, color = "#000000", linestyle = "--", marker = "+",markersize = 4)
```

```python
plt.show()
```
