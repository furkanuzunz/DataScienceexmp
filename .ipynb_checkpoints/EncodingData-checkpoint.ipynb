```python
import pandas as pd
import seaborn as sns ##dataseti getireceğiz ondana aslında
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df[["sex","class","embark_town"]].isna().sum()
```

```python
df[["sex","class","embark_town"]]
```

```python
#embarkda 2 tane eksik veri oldugu icin 2 satırlık verriyi dropna edebilirim
#df["embark_town"].dropna()
df = df.dropna(subset=['embark_town'])
```

```python
df
```

```python
df[["sex","class","embark_town"]].isna().sum()
```

```python
##ONE HOT ENCODING
```

```python
df["sex"].value_counts()
```

```python
df["embark_town"].value_counts()
```

```python
df["class"].value_counts()
```

```python
df_onehot = pd.get_dummies(df,columns = ["sex","embark_town"],drop_first = True)
```

```python
df_onehot
```

```python
df_onehot.columns
```

```python
##LABEL ENCODER
```

```python
label_encoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label
```

```python
df_label["sex"] = label_encoder.fit_transform(df_label["sex"])
```

```python
df_label
```

```python
#ordinal encoding
```

```python
orth_class_order = ["Third","Second","First"]
```

```python
ordinal_encoder = OrdinalEncoder(categories=[orth_class_order])
```

```python
df_ordinal = df.copy()
```

```python
df_ordinal["class"] = ordinal_encoder.fit_transform(df_ordinal[["class"]]) 
```

```python
df_ordinal
```

```python
#visualization to see the effects of encoding
```

```python
fig,axes = plt.subplots(1,3,figsize = (15,5))
df_sex_counts = df["sex"].value_counts()
df_sex_counts.plot(kind = "bar",ax = axes[0],title = "Original Categorical")

df_label["sex"].value_counts().plot(kind = "bar",ax = axes[1],title = "Label Encoded")
df_onehot[["sex_male"]].value_counts().plot(kind = "bar",ax = axes[2],title = "One-Hot Encoded")
```
